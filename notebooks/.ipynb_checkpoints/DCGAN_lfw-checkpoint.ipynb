```python
import mxnet as mx
from mxnet import gluon, autograd, nd
from mxnet.gluon import nn, utils
from matplotlib import pyplot as plt
from datetime import datetime
import numpy as np
import matplotlib as mpl
import os, sys, tarfile, time, logging
```

```python
batch_size = 64
latent_z_size = 200

use_gpu = True
context = mx.gpu()

lr = .00002
beta1 = 0.5

data_path = 'lfw_dataset'
```

```python
with tarfile.open('lfw-deepfunneled.tgz') as tar:
        tar.extractall(path=data_path)
```

```python
target_wd = 64
target_ht = 64
img_list = []

def transform(data, target_wd, target_ht):
    data = mx.image.imresize(data, target_wd, target_ht) # resize to target_wd * target_ht
    data = nd.transpose(data, (2,0,1)) # (channel * height * width)
    data = data.astype(np.float32)/127.5 - 1     # normalize to [-1, 1]
    if data.shape[0] == 1:
        data = nd.tile(data, (3, 1, 1))
    return data.reshape((1,) + data.shape)

for path, _, fnames in os.walk(data_path):
    for fname in fnames:
        if not fname.endswith('.jpg'):
            continue
        img = os.path.join(path, fname)
        img_arr = mx.image.imread(img)
        img_arr = transform(img_arr, target_wd, target_ht)
        img_list.append(img_arr)

train_idx = np.random.choice(range(len(img_list)), size = np.int(len(img_list) * .7))

_tr_data = [x for i, x in enumerate(img_list) if i in train_idx]
_va_data = [x for i, x in enumerate(img_list) if i not in train_idx]

train_data = mx.io.NDArrayIter(data= nd.concat(*_tr_data, dim = 0).as_in_context(context) , batch_size=batch_size, shuffle = True)
valid_data = mx.io.NDArrayIter(data= nd.concat(*_va_data, dim = 0).as_in_context(context) , batch_size=batch_size, shuffle = True)
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#imgplot = plt.imshow(X[0][0] * 255,cmap='gray')
#plt.show()
```

```python
latent_z_size = 100

use_gpu = True
ctx = mx.gpu() if use_gpu else mx.cpu()

lr = 0.0002
beta1 = 0.5
```

```python
# build the generator
nc = 3
ngf = 32
netG = nn.Sequential()
with netG.name_scope():
    # input is Z, going into a convolution
    netG.add(nn.Conv2DTranspose(ngf * 8, 4, 1, 0, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*8) x 4 x 4
    netG.add(nn.Conv2DTranspose(ngf * 4, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*4) x 8 x 8
    netG.add(nn.Conv2DTranspose(ngf * 2, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*2) x 16 x 16
    netG.add(nn.Conv2DTranspose(ngf, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. () x 32 x 32
    netG.add(nn.Conv2DTranspose(nc, 4, 2, 1, use_bias=False))
    netG.add(nn.Activation('tanh'))
    # state size. (nc) x 64 x 64
```

```python
netG
```

```python
fixed_z = nd.random.uniform(shape = (16, latent_z_size, 1, 1), ctx = context)
netG.collect_params().initialize(mx.init.Xavier(), ctx = context)
netG(fixed_z)
```

```python
# build the discriminator
ndf = 32
netD = nn.Sequential()
with netD.name_scope():
    # input is (nc) x 64 x 64
    netD.add(nn.Conv2D(ndf, 4, 2, 1, use_bias=False))
    netD.add(nn.LeakyReLU(0.2))
    # state size. (ndf) x 32 x 32
    netD.add(nn.Conv2D(ndf * 2, 4, 2, 1, use_bias=False))
    netD.add(nn.BatchNorm())
    netD.add(nn.LeakyReLU(0.2))
    # state size. (ndf*2) x 16 x 16
    netD.add(nn.Conv2D(ndf * 4, 4, 2, 1, use_bias=False))
    netD.add(nn.BatchNorm())
    netD.add(nn.LeakyReLU(0.2))
    # state size. (ndf*4) x 8 x 8
    netD.add(nn.Conv2D(ndf * 8, 4, 2, 1, use_bias=False))
    netD.add(nn.BatchNorm())
    netD.add(nn.LeakyReLU(0.2))
    # state size. (ndf*8) x 4 x 4
    netD.add(nn.Conv2D(1, 4, 1, 0, use_bias=False))
    # state size. (ndf) x 1 x 1
```

```python
netD
```

```python
# loss
loss = gluon.loss.SigmoidBinaryCrossEntropyLoss()
# initialize the generator and the discriminator
netG.initialize(mx.init.Normal(0.02), ctx=context)
netD.initialize(mx.init.Normal(0.02), ctx=context)

# trainer for the generator and the discriminator
trainerG = gluon.Trainer(netG.collect_params(), 'adam', {'learning_rate': lr, 'beta1': beta1})
trainerD = gluon.Trainer(netD.collect_params(), 'adam', {'learning_rate': lr, 'beta1': beta1})
```

```python
# test에서 항상 뽑아볼 수 있는 고정된 z를 사전에 뽑아둡니다.
fixed_z = nd.random.uniform(shape = (16, latent_z_size, 1, 1), ctx = context)
```

```python
epochs = 2000

from datetime import datetime
import time
import logging

real_label = nd.ones((batch_size,), ctx=context)
fake_label = nd.zeros((batch_size,),ctx=context)

def facc(label, pred):
    pred = pred.ravel()
    label = label.ravel()
    return ((pred > 0.5) == label).mean()
metric = mx.metric.CustomMetric(facc)

stamp =  datetime.now().strftime('%Y_%m_%d-%H_%M')
logging.basicConfig(level=logging.DEBUG)

for epoch in range(epochs):
    tic = time.time()
    btic = time.time()
    iter = 0
    train_data.reset()
    for i, batch in enumerate(train_data):
        data = batch.data[0].as_in_context(context)
        #print([x.shape for x in data])
        #X = np.zeros((batch_size, 28, 28))
        #for idx in range(data.shape[0]):
        #    X[idx] = data.asnumpy()[idx,0,:,:]
        #X = np.asarray([cv2.resize(x, (64,64)) for x in X])
        #X = X.reshape((batch_size,1,64,64))
        #data = nd.array(X,ctx=ctx)
        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################
        latent_z = mx.nd.random_normal(0, 1, shape=(batch_size, latent_z_size, 1, 1), ctx = context)
    
        #Gradient를 산출 및 이를 Back prop.에 적용
        with autograd.record():
            # train with real image
            output = netD(data).reshape((-1, 1))
            #print('output shape = {}'.format(output.shape))
            #print('real_label shape = {}'.format(real_label.shape))
            errD_real = loss(output, real_label)
            metric.update([real_label,], [output,])

            # train with fake image
            fake = netG(latent_z)
            output = netD(fake.detach()).reshape((-1, 1))
            errD_fake = loss(output, fake_label)
            errD = errD_real + errD_fake
            errD.backward()
            metric.update([fake_label,], [output,])

        trainerD.step(data.shape[0])

        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        with autograd.record():
            fake = netG(latent_z)
            output = netD(fake).reshape((-1, 1))
            errG = loss(output, real_label)
            errG.backward()

        trainerG.step(data.shape[0])

    # Print log infomation every ten batches
    if epoch % 10 == 0:
        name, acc = metric.get()
        logging.info('speed: {} samples/s'.format(batch_size / (time.time() - btic)))
        logging.info('discriminator loss = %f, generator loss = %f, binary training acc = %f at iter %d epoch %d'
                 %(nd.mean(errD).asscalar(),
                   nd.mean(errG).asscalar(), acc, iter, epoch))
        gen_img = netG(fixed_z)
        #print(gen_img.shape)
        plt.figure(figsize=(10,10))
        for i in range(8):
            plt.subplot(1,8,i+1)
            # 이미 64 * 64의 크기로 되어 있음
            plt.imshow(gen_img[i][0].asnumpy())
            plt.axis('off')
            plt.Annotation(i, (0, 0))
        plt.show()
    iter = iter + 1
    btic = time.time()

    name, acc = metric.get()
    metric.reset()
    # logging.info('\nbinary training acc at epoch %d: %s=%f' % (epoch, name, acc))
    # logging.info('time: %f' % (time.time() - tic))

    # Visualize one generated image for each epoch
```
