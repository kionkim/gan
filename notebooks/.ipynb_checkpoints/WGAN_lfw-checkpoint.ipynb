```python
import mxnet as mx
from mxnet import gluon, autograd, nd, metric
from mxnet.gluon import nn, utils
from matplotlib import pyplot as plt
from datetime import datetime
import numpy as np
import matplotlib as mpl
import os, sys, tarfile, time, logging
```

```python
batch_size = 64
latent_dim = 200

ctx = mx.gpu()

data_path = 'lfw_dataset'
```

```python
with tarfile.open('lfw-deepfunneled.tgz') as tar:
        tar.extractall(path=data_path)
```

```python
target_wd = 64
target_ht = 64
img_list = []

def transform(data, target_wd, target_ht):
    # resize to target_wd * target_ht
    data = mx.image.imresize(data, target_wd, target_ht)
    # transpose from (target_wd, target_ht, 3)
    # to (3, target_wd, target_ht)
    data = nd.transpose(data, (2,0,1))
    
    data = data.astype(np.float32)  /127.5 - 1 #  normalize to [-1, 1]
    # if image is greyscale, repeat 3 times to get RGB image.
    if data.shape[0] == 1:
        data = nd.tile(data, (3, 1, 1))
    return data.reshape((1,) + data.shape)

for path, _, fnames in os.walk(data_path):
    for fname in fnames:
        if not fname.endswith('.jpg'):
            continue
        img = os.path.join(path, fname)
        img_arr = mx.image.imread(img)
        img_arr = transform(img_arr, target_wd, target_ht)
        img_list.append(img_arr)
train_data = mx.io.NDArrayIter(data=nd.concatenate(img_list), batch_size=batch_size, shuffle = True)
```

```python
def  visualize(img_arr):
    plt.imshow(((img_arr.asnumpy().transpose(1, 2, 0) +1) * 127.5).astype(np.uint8))
    plt.axis('off')

for i, j in enumerate(np.random.choice(range(100), 3)):
    plt.subplot(1,3,i+1)
    visualize(img_list[j + 10][0])
plt.show()
```

```python
# build the generator
nc = 3
ngf = 32
netG = nn.Sequential()
with netG.name_scope():
    # input is Z, going into a convolution
    netG.add(nn.Conv2DTranspose(ngf * 8, 4, 1, 0, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*8) x 4 x 4
    netG.add(nn.Conv2DTranspose(ngf * 4, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*4) x 8 x 8
    netG.add(nn.Conv2DTranspose(ngf * 2, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. (ngf*2) x 16 x 16
    netG.add(nn.Conv2DTranspose(ngf, 4, 2, 1, use_bias=False))
    netG.add(nn.BatchNorm())
    netG.add(nn.Activation('relu'))
    # state size. () x 32 x 32
    netG.add(nn.Conv2DTranspose(nc, 4, 2, 1, use_bias=False))
    netG.add(nn.Activation('tanh'))
    # state size. (nc) x 64 x 64
```

```python
class Gen_mlp(gluon.HybridBlock):
    def __init__(self, **kwargs):
        super(Gen_mlp, self).__init__(**kwargs)
        with self.name_scope():
            self.dense1 = nn.Dense(512)
            self.dense2 = nn.Dense(512)
            self.dense3 = nn.Dense(512)
            self.dense4 = nn.Dense(3 * 64 * 64)
            
    def hybrid_forward(self, F, x):
        #print('input.shape = {}'.format(x.shape))
        x = self.dense1(x)
        x = nd.relu(x)
        x = self.dense2(x)
        x = nd.relu(x)
        x = self.dense3(x)
        x = nd.relu(x)
        x = self.dense4(x)
        x = nd.relu(x)
        x = nd.reshape(x, shape = (-1, 3, 64, 64))  
        return x
```

```python
class Critic(gluon.HybridBlock):
    def __init__(self, **kwargs):
        super(Critic, self).__init__(**kwargs)
        with self.name_scope():
            self.dense1 = nn.Dense(64)
            self.dense2 = nn.Dense(64)
            self.dense3 = nn.Dense(64)
            self.dense4 = nn.Dense(1)
            
    def hybrid_forward(self, F, x):
        x = self.dense1(x)
        x = nd.relu(x)
        x = self.dense2(x)
        x = nd.relu(x)
        x = self.dense3(x)
        x = nd.relu(x)
        x = self.dense4(x)   
        return x
```

```python
gen = Gen_mlp()
critic = Critic()
gen.collect_params().initialize(mx.init.Normal(0.02), ctx = mx.gpu())
critic.collect_params().initialize(mx.init.Normal(0.02), ctx = mx.gpu())
```

```python
gen_trainer = gluon.Trainer(gen.collect_params(), 'RMSProp', {'learning_rate': lr})
critic_trainer = gluon.Trainer(critic.collect_params(), 'RMSProp', {'learning_rate': lr})
```

```python
from tqdm import tqdm, tqdm_notebook
import matplotlib

for epoch in tqdm_notebook(range(epochs), desc = 'epoch'):
    tic = time.time()
    btic = time.time()
    train_data.reset()
    
    ############################
    # (1) Update Critique network: maximize E(f_w(X)) - E(f_w(G(z)))
    # 모든 sample을 이용해서 critique 함수를 극대화시킴
    ###########################
    for i, batch in enumerate(train_data):
        data = batch.data[0].as_in_context(ctx)
        latent_z = nd.random.normal(shape =(batch_size, latent_dim), ctx = ctx) # Latent data
        e_gen = 0
        e_dat = 0
        for j in range(n_critic_steps):
            with autograd.record():
                # We have to find supremum
                neg_crit = - (nd.mean(critic(nd.flatten(data))) - nd.mean(critic(nd.flatten(gen(latent_z)))))
                e_dat = nd.mean(critic(nd.flatten(data)))
                e_gen = nd.mean(critic(nd.flatten(gen(latent_z))))
                neg_crit.backward()
            critic_trainer.step(batch_size)
            [critic.collect_params()[p].set_data(nd.clip(critic.collect_params()[p].data(), -clip, clip)) for p in critic.collect_params()]            
        
        # zerorize gradient
        for param in critic.params:
            param.zero_grad()
        
        ############################
        # (2) Update G network: maximize g_theta(z)
        ###########################
        latent_z = nd.random.normal(shape = (batch_size, latent_dim), ctx =ctx)
        with autograd.record():
            gen_loss = - nd.mean(critic(gen(latent_z)))
            gen_loss.backward()
            #print('gen_loss = {}'.format(gen_loss.asnumpy()[0]))
        gen_trainer.step(batch.data[0].shape[0])
        
        # zerorize gradient
        for param in gen.params:
            param.zero_grad()
            
    print('-'* 30)    
    print('distance: {:.4f}, f_w_data: {:.4f}, f_w_gen: {:.4f}, gen_loss: {:.4f}'.format(\
                (-neg_crit).asnumpy()[0],\
                e_dat.asnumpy()[0], \
                e_gen.asnumpy()[0], \
                nd.mean(critic(gen(latent_z))).asnumpy()[0]                                          
            ))
    print('-'* 30)

    
    #save dis/gen loss
    if epoch % log_freq == 0:
        tqdm.write('speed: {} seconds'.format(time.time() - btic))
        tqdm.write('distance = %f at epoch %d'%(distance[0].asnumpy()[0], epoch))
        btic = time.time()
        iter_loss['crit'].append(nd.mean(distance).asscalar())
        iter_loss['gen'].append(nd.mean(gen_loss).asscalar())
        z = nd.random.normal(shape = (1, latent_dim), ctx = ctx)
        #print('z shape = {}'.format(z.shape))
        gen_img = gen(z)
        #print('z shape = {}'.format(gen_img.shape))
        _norm_img = ((np.squeeze(gen_img.asnumpy(), axis = 0)\
                      .transpose(1, 2, 0) + 1) * 127.5)\
                      .astype(np.uint8)
        plt.imshow(_norm_img) #Needs to be in row,col order
        plt.savefig('../output/test_{}.png'.format(epoch))
```
